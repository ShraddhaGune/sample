```python
class Student:
    def __init__(self, roll_no, name, course):
        self.roll_no = roll_no
        self.name = name
        self.course = course

    def display(self):
        print(f"Roll No: {self.roll_no}, Name: {self.name}, Course: {self.course}")

class StudentDatabase:
    def __init__(self):
        #empty list to add student object
        self.students = []

    def accept(self, student):
        #adding student object to empty list
        self.students.append(student)
        print("Student added successfully.")

    def delete(self, roll_no):
        for student in self.students:
            if student.roll_no == roll_no:
                self.students.remove(student)
                print("Student deleted successfully.")
                return
        print("Student not found.")

    def search(self, roll_no):
        for student in self.students:
            if student.roll_no == roll_no:
                student.display()
                return
        print("Student not found.")

    def display_all(self):
        for student in self.students:
            student.display()

    def update(self, roll_no, name=None, course=None):
        for student in self.students:
            if student.roll_no == roll_no:
                if name:
                    student.name = name
                if course:
                    student.course = course
                print("Student updated successfully.")
                return
        print("Student not found.")

# Example 
r = int(input("Enter roll number"))
n = input("Enter name")
c = input("Enter course taken")
db = StudentDatabase()
one = Student(101, "John Doe", "Computer Science")
two = Student(102, "Jane Doe", "Mechanical Engineering")
three = Student(r,n,c)
db.accept(one)
db.accept(two)
db.accept(three)
db.display_all()

db.update(101, name="John Smith")
db.search(101)

db.delete(102)
db.display_all()
```
